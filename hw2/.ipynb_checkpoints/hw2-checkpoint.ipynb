## Impurity Measures

Impurity is a measure of how often a randomly chosen element from the set would be incorrectly labeled if it was randomly labeled according to the distribution of labels in the subset. Implement the functions `calc_gini` and `calc_entropy`. You are encouraged to test your implementation (10 points).

````python
def get_labels_probabilities(data):
    labels = data[:, -1]
    total_num_of_instances = len(labels)
    probabilities = np.unique(labels, return_counts=True)[1]/total_num_of_instances
    return probabilities
````

````python
def calc_gini(data):
    """
    Calculate gini impurity measure of a dataset.
 
    Input:
    - data: any dataset where the last column holds the labels.
 
    Returns the gini impurity.    
    """
    gini = 0.0
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################
    
    probabilities = get_labels_probabilities(data)
    gini = 1 - (probabilities**2).sum()
    
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return gini
````

````python
def calc_entropy(data):
    """
    Calculate the entropy of a dataset.

    Input:
    - data: any dataset where the last column holds the labels.

    Returns the entropy of the dataset.    
    """
    entropy = 0.0
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################
    
    probabilities = get_labels_probabilities(data)
    entropy = -(probabilities*np.log2(probabilities)).sum()
    
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return entropy
````

## Goodness of Split

Given a feature the Goodnees of Split measures the reduction in the impurity if we split the data according to the feature.
$$
\Delta\varphi(S, A) = \varphi(S) - \sum_{v\in Values(A)} \frac{|S_v|}{|S|}\varphi(S_v)
$$

In our implementation the goodness_of_split function will return either the Goodness of Split or the Gain Ratio as learned in class. You'll control the return value with the `gain_ratio` parameter. If this parameter will set to False (the default value) it will return the regular Goodness of Split. If it will set to True it will return the Gain Ratio.
$$
GainRatio(S,A)=\frac{InformationGain(S,A)}{SplitInformation(S,A)}
$$
Where:
$$
InformationGain(S,A)=Goodness\ of\ Split\ calculated\ with\ Entropy\ as\ the\ Impurity\ function \\
SplitInformation(S,A)=- \sum_{a\in A} \frac{|S_a|}{|S|}\log\frac{|S_a|}{|S|}
$$
NOTE: you can add more parameters to the function and you can also add more returning variables (The given parameters and the given returning variable should not be touch). (10 Points)

````python
def goodness_of_split(data, feature, impurity_func, gain_ratio=False):
    """
    Calculate the goodness of split of a dataset given a feature and impurity function.

    Input:
    - data: any dataset where the last column holds the labels.
    - feature: the feature index.
    - impurity func: a function that calculates the impurity.
    - gain_ratio: goodness of split or gain ratio flag.

    Returns the goodness of split (or the Gain Ration).  
    """
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################
    
    if gain_ratio == True or impurity_func == calc_entropy:
        impurity = calc_entropy(data)
    else:
        impurity = impurity_func(data)
    
    impurity_after = 0
    
    feature_values_count_dict = dict(zip(*np.unique(data[:,feature], return_counts=True)))
    total_number_of_samples = data.shape[0]
    
    for value in feature_values_count_dict.keys():
        splited_data = data[data[:,feature] == value]
        impurity_after += (feature_values_count_dict[value]/total_number_of_samples)*impurity_func(splited_data)
    
    if gain_ratio == True:
        information_gain = impurity - impurity_after
        proportions = np.array(list(feature_values_count_dict.values()))/total_number_of_samples
        split_in_information = -(proportions*np.log2(proportions)).sum()
        gain_ratio = information_gain/split_in_information
        goodness = gain_ratio
    else:
        goodness = impurity - impurity_after
        
    
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return goodness    
````

````python
class DecisionNode:
    """
    This class will hold everything you require to construct a decision tree.
    The structure of this class is up to you. However, you need to support basic 
    functionality as described above. It is highly recommended that you 
    first read and understand the entire exercise before diving into this class.
    """
    def __init__(self, value, feature,  parent_feature):
        self.value = value
        self.feature = feature # column index of criteria being tested
        self.parent_feature = parent_feature
        self.children = []
        self.data = None
        self.is_leaf = False
        self.pred = None
        
        
    def add_child(self, node):
        self.children.append(node)
        
    
    def add_data(self, data):
        self.data = data
        unique, counts = np.unique(data[:, -1], return_counts=True)
        self.class_counts = dict(zip(unique, counts))
````

````python
def split_node(data, impurity, parent_feature, gain_ratio=False, max_depth=1000,
               feature_indicies=None, value='', min_samples_split=1):
    
    features_goodness_of_split_dict = {}
    
    for i in feature_indicies:
        features_goodness_of_split_dict[i] = goodness_of_split(data, i, impurity, gain_ratio)
    
    best_feature_to_split_by = max(features_goodness_of_split_dict, key=features_goodness_of_split_dict.get)
    
    node = DecisionNode(value, best_feature_to_split_by, parent_feature)
    
    feature_values = np.unique(data[:, best_feature_to_split_by])
    if features_goodness_of_split_dict[best_feature_to_split_by] == 0 or not max_depth or data.shape[0] <= min_samples_split:
        child_node = DecisionNode(value, None, parent_feature)
        child_node.is_leaf = True
        child_node.add_data(data)
        labels, labels_count = np.unique(data[:,-1], return_counts=True)
        child_node.pred = labels[np.argmax(labels_count)]
        
        return child_node
    
    for feature_value in feature_values:
        child_data = data[data[:, best_feature_to_split_by] == feature_value]
        values, count = np.unique(child_data[:, -1], return_counts=True)
        parent_feature = best_feature_to_split_by
    
        
        if len(count) == 1:
            child_node = DecisionNode(feature_value, None, parent_feature)
            child_node.is_leaf = True
            child_node.pred = values[0]
        else:
            child_node = split_node(child_data, impurity, parent_feature, gain_ratio, max_depth-1,
                                    feature_indicies, feature_value, min_samples_split)
            
    
        child_node.add_data(child_data)
        node.add_child(child_node)
    
    return node


def build_tree(data, impurity, gain_ratio=False, min_samples_split=1, max_depth=1000, feature_indicies=None):
    """
    Build a tree using the given impurity measure and training dataset. 
    You are required to fully grow the tree until all leaves are pure. 

    Input:
    - data: the training dataset.
    - impurity: the chosen impurity measure. Notice that you can send a function
                as an argument in python.
    - gain_ratio: goodness of split or gain ratio flag
    - min_samples_split: the minimum number of samples required to split an internal node
    - max_depth: the allowable depth of the tree

    Output: the root node of the tree.
    """
    root = None
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################
    
    if len(data) == 1:
        return data[0][-1]
    
    if feature_indicies is None:
        feature_indicies = np.arange(data.shape[1] - 1)
        
    value = 'ROOT'
    parent_feature = -1
    
    root = split_node(data, impurity, parent_feature, gain_ratio, max_depth, feature_indicies, value, min_samples_split)
    root.add_data(data)
    
    
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return root
````

````python
def predict(node, instance):
    """
    Predict a given instance using the decision tree
 
    Input:
    - root: the root of the decision tree.
    - instance: an row vector from the dataset. Note that the last element 
                of this vector is the label of the instance.
 
    Output: the prediction of the instance.
    """
    pred = None
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################

    while (node.is_leaf == False):
        split_by_feature = node.feature
        instance_feature_value = instance[split_by_feature]
        
        next_step = -1
        
        for ind, child in enumerate(node.children):
            if child.value == instance_feature_value:
                next_step = ind
                
        
        if next_step != -1:
            node = node.children[next_step]
        else:
            
            labels, labels_count = np.unique(node.data[:,-1], return_counts=True)
            node.pred = labels[np.argmax(labels_count)]
            return node.pred
            
    
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return node.pred
````

````python
def calc_accuracy(node, dataset):
    """
    Predict a given dataset using the decision tree
 
    Input:
    - node: a node in the decision tree.
    - dataset: the dataset on which the accuracy is evaluated
 
    Output: the accuracy of the decision tree on the given dataset (%).
    """
    accuracy = 0
    ###########################################################################
    # TODO: Implement the function.                                           #
    ###########################################################################
    
    is_correct = np.array([])
    
    for sample in dataset:
        label = sample[-1]
        prediction = predict(node, sample)
        
        is_correct = np.append(is_correct, label==prediction)
    
    accuracy = is_correct.sum()/dataset.shape[0]
        
    
    ###########################################################################
    #                             END OF YOUR CODE                            #
    ###########################################################################
    return accuracy 
````
